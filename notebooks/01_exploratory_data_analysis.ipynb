## 1. Library Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Paths & Column Names

```python
RAW_TRAIN_PATH = "../data/raw/KDDTrain+.txt"

columns = [
    "duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
    "wrong_fragment","urgent","hot","num_failed_logins","logged_in","num_compromised",
    "root_shell","su_attempted","num_root","num_file_creations","num_shells",
    "num_access_files","num_outbound_cmds","is_host_login","is_guest_login",
    "count","srv_count","serror_rate","srv_serror_rate","rerror_rate",
    "srv_rerror_rate","same_srv_rate","diff_srv_rate","srv_diff_host_rate",
    "dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
    "dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate",
    "dst_host_srv_serror_rate","dst_host_rerror_rate",
    "dst_host_srv_rerror_rate","label"
]
```

## 3. Load Data

```python
df = pd.read_csv(RAW_TRAIN_PATH, header=None, names=columns)
```

## 4. Basic Info

```python
print("Shape of dataset:", df.shape)
print("\nFirst 5 rows:")
display(df.head())
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
```

## 5. Class Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(y='label', data=df, order=df['label'].value_counts().index)
plt.title("Class Distribution (Training Data)")
plt.xlabel("Count")
plt.ylabel("Class Label")
plt.show()

print("\nClass Distribution Table:")
print(df['label'].value_counts())
```

## 6. Numerical Feature Distribution

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
sample_numeric_cols = numeric_cols[:5]  # just first 5 for quick view

df[sample_numeric_cols].hist(figsize=(12,8), bins=20)
plt.suptitle("Sample Numerical Feature Distributions")
plt.show()
```

## 7. Correlation Heatmap

```python
plt.figure(figsize=(12,10))
sns.heatmap(df[numeric_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

## 8. Protocol, Service, Flag Frequencies

```python
cat_cols = ["protocol_type", "service", "flag"]

for col in cat_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f"{col.capitalize()} Frequency")
    plt.show()
```
